```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
def getTickerPrice(ticker:str,date)->float:
    return random.uniform(0, 100)
def validate_dataset(df):
    if df.empty:
        return False
    return True
def fill_missing_values(df):
    df['Size'].fillna(1, inplace=True)
def date_sort(df):
    df_sorted = df.sort_values(by='Date')
    df_sorted.reset_index(drop=True, inplace=True)
    return df_sorted
def Symbol_sort(df):
    df_sorted = df.sort_values(by='Symbol')
    df_sorted.reset_index(drop=True, inplace=True)
    return df_sorted
def total_profit(df):
    profit=0
    for idx, row in df.iterrows():
        if row['Side']=='buy':
            profit-=row['Price']*row['Size']
        else:
            profit+=row['Price']*row['Size']
    return profit
def drawdown(df):
    drawdown=0
    current_holding={}
    total_invested=0
    for idx, row in df.iterrows():
        if row['Symbol'] not in current_holding:
            if row['Side']=='buy':
                current_holding[row['Symbol']]=[row['Price']*row['size'],row['Size']]
            else:
                current_holding[row['Symbol']]=[row['Price']*row['size'],-row['Size']]
            total_invested+=row['Price']*row['Size']
        else:
            if row['Side']=='buy':
                if current_holding[row['Symbol']][1]>0:
                    current_holding[row['Symbol']][0]+=row['Price']*row['Size']
                    current_holding[row['Symbol']][1]+=row['Size']
                    total_invested+=row['Price']*row['Size']
                else:
                    temp=current_holding[row['Symbol']][0]
                    current_holding[row['Symbol']][0]=abs(row['Price']*row['Size']-current_holding[row['Symbol']][0])
                    current_holding[row['Symbol']][1]+=row['Size']
                    total_invested=total_invested-temp+current_holding[row['Symbol']][0]
            else:
                if current_holding[row['Symbol']][1]<0:
                    current_holding[row['Symbol']][0]+=row['Price']*row['Size']
                    current_holding[row['Symbol']][1]-=row['Size']
                    total_invested+=row['Price']*row['Size']
                else:
                    temp=current_holding[row['Symbol']][0]
                    current_holding[row['Symbol']][0]=abs(row['Price']*row['Size']-current_holding[row['Symbol']][0])
                    current_holding[row['Symbol']][1]-=row['Size']
                    total_invested=total_invested-temp+current_holding[row['Symbol']][0]
        if current_holding[row['Symbol']][1]==0:
            del current_holding[row['Symbol']]
            
        end_date=df.iloc[idx+1]['Date']
        start_date=row['Date']
        price_store={}
        temp_profit=0
        while start_date <= end_date:
            for key in current_holding:
                share_price=getTickerPrice(key,start_date)
                price_store[key]=share_price
            for key in current_holding:
                if current_holding[key][1]>0:
                    temp_profit+=(price_store[key]-current_holding[key][0])*current_holding[key][1]
                else:
                    temp_profit+=(current_holding[key][0]-price_store[key])*current_holding[key][1]
            drawdown=min(drawdown,temp_profit)

            start_date+= timedelta(days=1)
        return drawdown
    


                
        
        
        

    

    
                


        

    

            
                    
                
            


```
